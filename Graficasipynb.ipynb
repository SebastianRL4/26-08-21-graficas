```python
#Ejemplo de graficas en python

import matplotlib.pyplot as plt
import numpy as np

def f1(x):
  return -1.00*(x**2)+3*x  #(-x^2)+3x

t1= np.arange (-10.00, 15.00, 0.1)

plt.plot(t1, f1(t1))
plt.grid(color='k', linestyle='-', linewidth=2)
plt.show()
```
